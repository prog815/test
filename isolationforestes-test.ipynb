# Библиотеки подтягиваем

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

# Загружаем данные

```python
data = pd.read_csv("../input/creditcardfraud/creditcard.csv")
data
```

# Смотрим на данные

```python
data.info()
```

```python
data.describe().T
```

# выделяем X и y

```python
X = data.drop(['Time','Class'],axis=1).values
y = data['Class'].astype(bool).values
```

# Разделяем обучение и тест

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_train.mean(), y_test.mean()
```

# создаем и учим модель

```python
model = IsolationForest(contamination=y_train.mean()) # долю выбросов определим среднему на обучающей выборке
model.fit(X_train)
```

# Предсказываем

```python
conv_pred = lambda pred : pd.Series(pred).apply(lambda x: 1 if x == -1 else 0).values
```

```python
pred_y_train = conv_pred(model.predict(X_train))
pred_y_test = conv_pred(model.predict(X_test))
```

# Качество в train

```python
print(classification_report(y_train,pred_y_train))
```

На обучении модель отлавливает 27% выбросов

# Качество в test

```python
print(classification_report(y_test,pred_y_test))
```

На тесте модель отлавливает 29% выбросов
